```python
def masks_per_person(year):
    START_YEAR = 2019
    year_delta = year - START_YEAR
    return 2 ** year_delta

WORLD_POP_2020 = 7_800_000_000
MASK_MASS_KG = .15
EARTH_MASS_KG = 5.972e24
JUPITER_MASS_KG = 1.898e27
SUN_MASS_KG = 1.989e30

def stats(year):
    masks_per_indiv = masks_per_person(year)
    total_masks = masks_per_indiv * WORLD_POP_2020
    total_mask_mass = total_masks * MASK_MASS_KG
    return (masks_per_indiv, total_mask_mass, EARTH_MASS_KG, JUPITER_MASS_KG, SUN_MASS_KG)
```

```python
import matplotlib.pyplot as plt
import numpy as np

YEARS = [year for year in range(2020, 2101)]

def do_plot(rows, years):
    plt.plot(years, rows)
    plt.xlabel("Year")
    plt.yscale("log")
    plt.title("MaskLOL")
    plt.legend(["Masks Per Person", 
                "Total Mask Mass (kg)", 
                "Earth Mass (kg)", 
                "Jupiter Mass (kg)", 
                "Sun Mass (kg)"])
    plt.show()
    
do_plot([stats(year) for year in YEARS], YEARS)
```
